## 1º Exercício: Conversão e Filtragem de Dataframe
- Converta as datas e horas do dataframe abaixo para o formato correto de data;
- Mescle a data e a hora em uma coluna única chamada de *data_hora*;
- Crie um filtro da coluna *data_hora* que filtre uma data em específico.

```python
# Dataframe proposto:
import pandas as pd
data = {
    'data': ['2024-06-01', '2024-06-02', '2024-06-03', '2024-06-04'],
    'hora': ['08:00:00', '09:30:00', '11:15:00', '14:45:00']}
df = pd.DataFrame(data)
display(df)
```

```python
# Aqui você deve converter a coluna data e hora para o formato correto.
import time
from datetime import datetime, timedelta
df['data'] = pd.to_datetime(df['data'])
data_format=df['data'].dt.strftime('%d/%m/%Y')
df['hora'] = pd.to_datetime(df['hora'], format='%H:%M:%S').dt.time
df['data'] = data_format
display(df)
df.dtypes
```

```python
# Aqui você deve combinar a data e hora para crirar a coluna "data_hora"

df['data'] = pd.to_datetime(df['data']) #Lucas precisei devolver a formatação antiga pra conseguir funcionar rsrs (pelo menos o Display ficou mais bonito)
df['data_hora'] = df.apply(lambda row: datetime.combine(row['data'], row['hora']), axis=1)
df['data_hora']
df['data'] = data_format #volta pra formatação data BR
```

```python
# Crie um filtro para a coluna data_hora que separe 50% dos dados
# (Por exemplo, o seu filtro de mostrar 2 linhas segundo o filtro da coluna data_hora)

df[df['data_hora'] >= datetime(2024,3,6)]
```
