## Desafio 1

Crie um dicionario com a estrutura semelhante ao tópico Revisão "Introdução ao Pandas", com as 3 colunas que existem nele: "Nome","Idade", e "Data_Nascimento" (Você não deve copiar os dados, você devem alterá-los)

Usando pandas, criar um DataFrame a partir de um dicionário de dados e adicionar uma coluna 'Dobro'.

A coluna nova "Dobro" deverá conter o dobro de valor da coluna "Idade" (exemplo: Idade 10, Dobro 20)

Ao final, você utilizando essa função deve retornar (return) esse dataframe novo.

```python
import pandas as pd

def criar_dataframe_e_dobro(dados):
    df = pd.DataFrame(dados)
    df['Dobro'] = df['Idade'] * 2
    return df

dados = {
    'Nome': ['Lionel Messi', 'Cristiano Ronaldo', 'Luka Modrić', 'Ronaldinho Gaúcho', 'Kaká'],
    'Idade': [35, 37, 36, 42, 41],
    'Data_Nascimento': ['1987-06-24', '1985-02-05', '1985-09-09', '1980-03-21', '1982-04-22']
}

df_novo = criar_dataframe_e_dobro(dados)
print(df_novo)
```

## Desafio 2

Apartir do dicionário "data" crie um dataframe chamado de "df1".

Converta os dados de "df1" para essas datas para datetime.

Crie o objeto "df2" que deverá conter valores de "df1" com a coluna "Valor" menor do que 400.

Crie o objetivo "df3" que deverá conter valores de "df1" maiores ou iguais a 400 da coluna "Valor".

```python
import pandas as pd

data = {
    'ID': [1, 2, 3, 4, 5],
    'Data': ['2024-03-01', '2024-03-05', '2024-03-10', '2024-03-15', '2024-03-20'],
    'Valor': [100, 200, 300, 400, 500]
}

df1 = pd.DataFrame(data)
df1['Data'] = pd.to_datetime(df1['Data'])
df2 = df1[df1['Valor'] < 400]
df3 = df1[df1['Valor'] >= 400]

print("DF1")
print(df1)
print("DF2")
print(df2)
print("DF3")
print(df3)
```

## Desafio 3

Crie um loop que a cada novo minuto ele devera informar qual o minuto atual.

O gatilho dessa lógica tem que ser acionado no inicio do minuto! ou seja, não deve esperar ou "demorar" segundos desnecessários.

```python
import time
from datetime import datetime

def minuto_atual():
    while True:
        hora_atual = datetime.now()
        minuto_atual = hora_atual.minute
        saldo_segundos = 60 - hora_atual.second
        print(f"Minuto atual: {minuto_atual:02d}")
        print(f"Segundos até o próximo minuto:", end=" ")
        for i in range(int(saldo_segundos)):
            print(f"{int(saldo_segundos) - i}", end=" ")
            time.sleep(1)
        print()

minuto_atual()
```
