## Desafio 1: Conversão de Colunas para Datetime

Você tem um DataFrame com colunas de datas no formato string. Sua tarefa é converter essas colunas para o formato datetime. Além disso, você deve extrair o mês e o dia da semana de cada data e adicioná-los como novas colunas no DataFrame.

### Tarefa:
- Converta as colunas 'Data_Inicial' e 'Data_Final' para datetime.
- Extraia o mês de 'Data_Inicial' e adicione como uma nova coluna 'Mes_Inicial'.
- Extraia o dia da semana de 'Data_Final' e adicione como uma nova coluna 'Dia_Semana_Final'.



```python
import pandas as pd
from datetime import datetime, timedelta

data = {
    'Evento': ['Evento A', 'Evento B', 'Evento C', 'Evento D'],
    'Data_Inicial': ['2023-06-21', '2023-07-01', '2023-07-15', '2023-08-05'],
    'Data_Final': ['2023-06-25', '2023-07-05', '2023-07-18', '2023-08-10']
}

df1 = pd.DataFrame(data)
print("DataFrame Original:")
display(df1)
```

```python
# Converta as colunas 'Data_Inicial' e 'Data_Final' para datetime.
df1['Data_Inicial'] = pd.to_datetime(df1['Data_Inicial'])
df1['Data_Final'] = pd.to_datetime(df1['Data_Final'])
print("DataFrame após função Datetime:")
display(df1)
df1.dtypes
```

```python
# Extraia o mês de 'Data_Inicial' e adicione como uma nova coluna 'Mes_Inicial'.
df1['Mes_Inicial'] = df1['Data_Inicial'].dt.month
display(df1)
```

```python
# Extraia o dia da semana de 'Data_Final' e adicione como uma nova coluna 'Dia_Semana_Final'.
# Dicionário para traduzir os dias da semana
dias_semana = {
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'}

df1['Dia_Semana_Final'] = df1['Data_Final'].dt.day_name().map(dias_semana)
display(df1)
```

## Desafio 2: Alteração de Valores Baseada em Datas

Você tem um DataFrame com datas e valores. Sua tarefa é alterar os valores de uma coluna com base em condições específicas de data.

### Tarefas
- Converta a coluna 'Data' para datetime.
- Modifique o valor da coluna 'Valor' para 999 se a data for igual a '2024-03-10'.
- Modifique o valor da coluna 'Valor' para 555 se a data estiver entre '2024-03-05' e '2024-03-15' (inclusive).

```python
data = {
    'ID': [1, 2, 3, 4, 5],
    'Data': ['2024-03-01', '2024-03-05', '2024-03-10', '2024-03-15', '2024-03-20'],
    'Valor': [100, 200, 300, 400, 500]
}

df2 = pd.DataFrame(data)
print("DataFrame Original:")
display(df2)
```

```python
# Converta a coluna 'Data' para datetime.
df2['Data'] = pd.to_datetime(df2['Data'])
print("DataFrame após função Datetime:")
display(df2)
df2.dtypes
```

```python
# Modifique o valor da coluna 'Valor' para 999 se a data for igual a '2024-03-10'
df2.loc[df2['Data'] == '2024-03-10', 'Valor'] = 999
display(df2)
```

```python
# Modifique o valor da coluna 'Valor' para 555 se a data estiver entre '2024-03-05' e '2024-03-15' (inclusive).
df2.loc[(df2['Data'] >= '2024-03-05') & (df2['Data'] <= '2024-03-15'), 'Valor'] = 555
display(df2)
```
