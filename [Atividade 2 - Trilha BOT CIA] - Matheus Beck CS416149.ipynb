# Exercício de Função: SOMA condicional

```python
# Agora sua vez! Aqui escreva uma função que recebe uma lista de valores e que só deve somar os valores ímpares:

num = list()
impar = 0
while True:
  n = int(input('Digite um número: '))
  num.append(n)
  r = str(input('Continuar? [S/N]')).strip().upper()[0]
  if n % 2 == 1:
    impar += n
  if r == 'N':
    break
print('=-'*30)
print(f'Você digitou os valores {num}')
print(f'A soma dos valores ímpares é {impar}')
```

# Exercício de Função: Elementos da Lista

```python
# Agora sua vez! Aqui escreva uma função que dada uma lista deve informar o menor valor, maior valor, valor médio, ...
#  ... e a contagem dos elementos dentro da lista

num = list()
maior = menor = med = cont = 0
while True:
  n = int(input('Digite um número: '))
  num.append(n)
  med += n
  r = str(input('Continuar? [S/N]')).strip().upper()[0]
  if cont == 0:
    maior = n
    menor = n
  cont += 1
  if n > maior:
    maior = n
  if n < menor:
    menor = n
  if r == 'N':
    break
print('=-'*30)
print(f'Os valores digitados foram {num}')
print(f'Você digitou um total de {cont} valores')
print(f'Dentre eles o maior valor digitado foi {maior}')
print(f'Dentre eles o menor valor digitado foi {menor}')
print(f'A média dos valores digitado é {med/cont}')
```
