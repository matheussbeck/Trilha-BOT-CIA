## 1º Exercício: Crieção de um dataframe, ele deve conte:
- 1 coluna "ID", que deverá conter numeros inteiros;<br>
- 1 coluna "Nome", que deverá conter nomes aleatórios;<br>
- 2 colunas "Vendas" e "Quantidade" com numeros inteiros;<br>
- Dataframe deve conter 6 linhas;<br>
- Após a criação do seu Dataframe faça operações de soma, minima, máximia nas colunas "Vendas" e "Quantidade";<br>
- Ainda com seu Dataframe, insira uma nova coluna chamada "Cidade" que deverá conter nomes de cidades;<br>
- Ao final dê um display no seu dataframe.

```python
# Crie o seu dataframe aqui:
import pandas as pd

ids = list()
dados = {'ID':ids,'Nome': ['Ana', 'Marina', 'Marco Antonio', 'Manuela', 'Joaquim', 'Genivaldo'],
         'Quantidade':[1 ,2 ,5 ,1 ,3 ,1 ],
         'Valor': [50.00 ,38.50 ,27.99 ,32.00 ,50.50 ,66.99 ]
         }
for c in range (0, len(dados['Nome'])):
  ids.append(c+1)
  c += 1
df = pd.DataFrame(dados)
display(df)
```

```python
# Faça operações com seu dataframe nas colunas "Vendas" e "Quantidade" aqui:
faturamento = list()
for i in range(len(dados['Quantidade'])):
  fat = dados['Quantidade'][i] * dados['Valor'][i]
  faturamento.append(fat)
dados['Faturamento'] = faturamento
df = pd.DataFrame(dados)
display(df)

print('-'*30)
print('Soma do Faturamento total')
print(f'R$ {df["Faturamento"].sum()}')
print('-'*30)

print('O maior valor de Faturamento')
df_max = df.sort_values(by='Faturamento', ascending=False)
display(df_max.loc[df_max['Faturamento'].idxmax()])
print('-'*30)

print('O Menor valor de Faturamento')
df_min = df.sort_values(by='Faturamento', ascending=True)
display(df_min.loc[df_min['Faturamento'].idxmin()])
print('-'*30)
```

```python
# Adicione a Coluna Cidade aqui:
dados['Cidade'] = ['Rio de Janeiro', 'Blumenau', 'Presidente Prudente', 'Curitiba', 'Londrina', 'Florianópolis']
```

```python
# Ao final, de um Display no seu dataframe aqui:
df = pd.DataFrame(dados)
display(df)
```

## 2º Exercício: Manipulações

Utilizando o dataframe criado abaixo, faça:

- Calcule a soma do "Custo" agrupado por "Localização";<br>
- Calcule a soma de "Tempo (horas)" agrupado por "Operação";<br>
- Mude os Nomes das fazendas (Dica: Utilize a função 'map' do pandas);<br>

> **Nota:** Neste material, você não aprendeu a utilizar a função 'map'. A ideia é que você aprenda a utilizá-la sozinho. A cadeira BOT CIA e Programação no geral exige muito autodidatismo e improviso.

```python
data = {
    'Data': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05',
             '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10'],
    'Operação': ['Plantio', 'Irrigação', 'Colheita', 'Transporte', 'Processamento',
                 'Plantio', 'Irrigação', 'Colheita', 'Transporte', 'Processamento'],
    'Quantidade (toneladas)': [100, 200, 300, 150, 250, 120, 220, 320, 180, 270],
    'Localização': ['Fazenda A', 'Fazenda B', 'Fazenda C', 'Fazenda A', 'Fazenda B',
                    'Fazenda C', 'Fazenda A', 'Fazenda B', 'Fazenda C', 'Fazenda A'],
    'Custo (R$)': [10000, 8000, 12000, 6000, 15000, 10500, 8500, 12500, 7000, 16000],
    'Tempo (horas)': [5, 3, 8, 4, 10, 6, 3.5, 8.5, 4.5, 11]
}
df = pd.DataFrame(data)
display(df)
```

```python
# Calcule a soma do "Custo" agrupado por "Localização"
df_custo = df.groupby('Localização')['Custo (R$)'].sum()
print(df_custo)
```

```python
# Calcule a soma de "Tempo (horas)" agrupado por "Operação"
df_tempo = df.groupby('Operação')['Tempo (horas)'].sum()
print(df_tempo)
```

```python
# Mude os Nomes das fazendas (Utilize a função 'map' do pandas)
mapa_fazendas = {
    'Fazenda A': 'Fazenda Genoveva',
    'Fazenda B': 'Fazenda Parintins',
    'Fazenda C': 'Fazenda Pé de Cabra'}
df['Localização'] = df['Localização'].map(mapa_fazendas)
```

```python
# Ao final, de um Display no seu dataframe aqui:
display(df)
```
